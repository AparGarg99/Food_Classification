# Re-arrange Dataset

```python
# create new directories
os.mkdir("food-101/Datasets"),os.mkdir("food-101/Datasets/train"),os.mkdir('food-101/Datasets/test')
```

```python
# create food class folders in directories
classes = open("/content/food-101/meta/classes.txt", "r").read().split()
for i in classes:
    os.mkdir('/content/food-101/Datasets/train/'+i)
    os.mkdir('/content/food-101/Datasets/test/'+i)
```

```python
# separate training data
train = open("/content/food-101/meta/train.txt", "r").read().split()
for i in train:
    im1 = PIL.Image.open("/content/food-101/images/"+i+".jpg")
    im1.save('/content/food-101/Datasets/train/'+i+".jpg") 
```

```python
# separate testing data
test = open("/content/food-101/meta/test.txt", "r").read().split()
for i in test:
    im1 = PIL.Image.open("/content/food-101/images/"+i+".jpg")
    im1.save('/content/food-101/Datasets/test/'+i+".jpg") 
```
